**CIE Coordinates (Simplified DIP Approach):**

Chromaticity coordinates can be simplified in DIP as:

- **x**: Red component
- **y**: Green component  
- **z**: Blue component (z = 1-x-y)

**Example**:  
780nm wavelength → x=0.74, y=0.26, z=0.00 (deep red)

---

**RGB Display Limitations:**

Modern displays reproduce only a **subset of human-visible colors**:

- **sRGB**: ~35% coverage
- **Adobe RGB**: ~50% coverage
- **DCI-P3**: ~45% coverage

---

**2D vs 3D Color Space:**

- **2D diagram**: Chromaticity (hue/saturation) at constant luminance
- **3D space**: Adds brightness dimension (Y-value)

### <a id='toc3_3_1_'></a>[Grayscale](#toc0_)

Converting any image to grayscale reduces color information to shades of gray.  
This simplification is useful for several reasons:

- **Data Reduction:**
  Only one channel is stored instead of three, reducing memory and computational requirements.

- **Preprocessing for Computer Vision:**
  Many algorithms (e.g., edge detection, segmentation, pattern recognition) work better or only on grayscale images.

- **Highlighting Intensity Information:**
  Luminance variations often correspond more closely to structural details than color variations.

- **Simplifying Analysis:**
  In medical imaging, document scanning, or texture analysis, grayscale helps isolate shape, contrast, and texture without color distractions.

**RGB to Grayscale Methods:**

1. **Luminance (Perceptual Weights, Standard Method) ✅:**  
   Human vision is more sensitive to green than red or blue.  
   Weighted sum formula:

   $$
   Y = 0.299 \, R + 0.587 \, G + 0.114 \, B
   $$

   - **Use when:** Accurate brightness representation is needed, e.g., human viewing, compression, perceptual tasks.

1. **Averaging Method:**  
   Arithmetic mean of the three channels:

   $$
   Y = \frac{R + G + B}{3}
   $$

   - **Use when:** Simplicity is more important than perceptual accuracy, e.g., quick visualization, basic preprocessing.

1. **L2 Norm (Euclidean Norm Method):**  
   Computes the magnitude of the RGB vector:

   $$
   Y = \sqrt{\frac{R^2 + G^2 + B^2}{3}}
   $$

   - **Use when:** Emphasizing overall intensity magnitude or treating RGB as a feature vector in scientific imaging.

1. **Max/Min Channel Method:**  
   Take the maximum or minimum channel value:

   $$
   Y = \max(R, G, B) \quad \text{or} \quad Y = \min(R, G, B)
   $$

   - **Use when:** Highlighting the strongest or weakest channel for edge detection or contrast emphasis.




#### <a id='toc3_3_1_1_'></a>[Manual](#toc0_)


```python
RGB_TO_GS = np.array(
    [
        [0.299, 0.587, 0.114],
    ],
    dtype=np.float32,
)

# approximation using pseudo-inverse [best linear least-squares approximation]
GS_TO_RGB = np.linalg.pinv(RGB_TO_GS)
```

```python
def rgb2gray(image: NDArray) -> NDArray[np.uint8]:
    img = (image.reshape(-1, 3) @ RGB_TO_GS.T).reshape(image.shape[:2])
    return np.clip(img, 0, 255).astype(np.uint8)


def gray2rgb(image: NDArray) -> NDArray[np.uint8]:
    img = (image.reshape(-1, 1) @ GS_TO_RGB.T).reshape(*image.shape, 3)
    img = ((img - img.min()) / np.ptp(img)) * 255
    return np.clip(img, 0, 255).astype(np.uint8)
```

```python
img_2_np_gs = rgb2gray(img_2_np)
img_2_np_rgb = gray2rgb(img_2_np_gs)

# plot
fig, axes = plt.subplots(1, 3, figsize=(10, 4), layout="compressed")
titles = ["Original", "RGB to GS", "GS to RGB (best approximation)"]
images = [img_2_np, img_2_np_gs, img_2_np_rgb]
for ax, img, title in zip(axes, images, titles):
    ax.imshow(img, cmap="gray" if img.ndim == 2 else None, vmin=0, vmax=255)
    ax.set_title(title)
    ax.axis("off")
plt.show()
```

#### <a id='toc3_3_1_3_'></a>[Comparison of Methods](#toc0_)


```python
R = img_2_np[:, :, 0].astype(np.float32)
G = img_2_np[:, :, 1].astype(np.float32)
B = img_2_np[:, :, 2].astype(np.float32)
```

```python
# 1. luminance (perceptual weights)
img_2_method_1 = 0.299 * R + 0.587 * G + 0.114 * B
img_2_method_1 = np.clip(img_2_method_1, 0, 255).astype(np.uint8)

# 2. averaging method
img_2_method_2 = (R + G + B) / 3
img_2_method_2 = np.clip(img_2_method_2, 0, 255).astype(np.uint8)

# 3. L2 norm (euclidean norm)
img_2_method_3 = np.sqrt((R**2 + G**2 + B**2) / 3)
img_2_method_3 = np.clip(img_2_method_3, 0, 255).astype(np.uint8)

# 4. max/min channel method
img_2_method_4_1 = np.max(np.dstack([R, G, B]), axis=2).astype(np.uint8)
img_2_method_4_2 = np.min(np.dstack([R, G, B]), axis=2).astype(np.uint8)
```

```python
# plot
fig, axes = plt.subplots(1, 6, figsize=(24, 4), layout="compressed")
titles = ["Original", "Perceptual Weights", "Averaging", "L2 Norm", "Max Channel", "Min Channel"]
images = [img_2_np, img_2_method_1, img_2_method_2, img_2_method_3, img_2_method_4_1, img_2_method_4_2]
for ax, img, title in zip(axes, images, titles):
    ax.imshow(img, cmap="gray" if img.ndim == 2 else None, vmin=0, vmax=255)
    ax.set_title(title)
    ax.axis("off")
plt.show()
```

### <a id='toc3_3_2_'></a>[RGB (Red, Green, Blue)](#toc0_)

- **Purpose:** Primary color space for digital displays
- **Usage:** Monitors, TVs, digital cameras, web design
- **Characteristics:** Additive color model

✍️ **Notes:**

OpenCV uses the BGR (Blue, Green, Red) channel order as its default color format instead of the more common RGB order for historical and compatibility reasons:

- **Historical Legacy:**  
  Early versions of OpenCV were designed to be compatible with the Windows bitmap (BMP) image format, which stores pixel data in BGR order. Using BGR natively simplified reading and writing BMP files without additional channel reordering.

- **Performance Considerations:**  
  By matching the underlying image formats and memory layouts commonly used on Windows systems, OpenCV avoids unnecessary data copying or channel swapping, which can improve performance.

- **Consistency in the OpenCV Ecosystem:**  
  Over time, many OpenCV functions, tutorials, and sample codes have been built around BGR images. Changing the default now would break backward compatibility and require widespread refactoring.

- **Interoperability:**  
  Since OpenCV often interfaces with other libraries and legacy code that expect BGR ordering, it maintains BGR as a practical default.

🔢 **RGB - BGR Conversion (Channel Swap):**

$$
\text{RGB → BGR}: 
\quad
\begin{bmatrix} B \\ G \\ R \end{bmatrix}
=
\begin{bmatrix} 0 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 0 & 0 \end{bmatrix}
\begin{bmatrix} R \\ G \\ B \end{bmatrix}
$$

$$
\text{BGR → RGB}:
\quad
\begin{bmatrix} R \\ G \\ B \end{bmatrix}
=
\begin{bmatrix} 0 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 0 & 0 \end{bmatrix}
\begin{bmatrix} B \\ G \\ R \end{bmatrix}
$$


#### <a id='toc3_3_2_1_'></a>[Manual](#toc0_)


```python
RGB_TO_BGR = np.array(
    [
        [0, 0, 1],
        [0, 1, 0],
        [1, 0, 0],
    ],
    dtype=np.float32,
)

BGR_TO_RGB = np.linalg.inv(RGB_TO_BGR)
```

```python
def rgb2bgr(image: NDArray) -> NDArray[np.uint8]:
    img = (image.reshape(-1, 3) @ RGB_TO_BGR.T).reshape(image.shape)
    return np.clip(img, 0, 255).astype(np.uint8)


def bgr2rgb(image: NDArray) -> NDArray[np.uint8]:
    img = (image.reshape(-1, 3) @ BGR_TO_RGB.T).reshape(image.shape)
    return np.clip(img, 0, 255).astype(np.uint8)
```

### <a id='toc3_3_3_'></a>[CMYK (Cyan, Magenta, Yellow, Key/Black)](#toc0_)

CMYK stands for **Cyan**, **Magenta**, **Yellow**, and **Key (Black)** — a subtractive color model used primarily in color printing.  
Unlike RGB (which is additive and designed for light-based displays), CMYK is based on ink or pigment absorption on physical media.

- **C (Cyan):** Absorbs red light.
- **M (Magenta):** Absorbs green light.
- **Y (Yellow):** Absorbs blue light.
- **K (Black):** Represents the black component used to enhance depth and reduce ink usage.

🥅 **Purpose of RGB to CMYK Conversion:**

- **Printing:**  
  CMYK is the standard color model for printers and press workflows, as it corresponds directly to ink usage.

- **Color Separation:**  
  The black (K) channel allows better shadow reproduction and text clarity without combining all three color inks.

- **Cost Efficiency:**  
  Using K instead of combining C, M, and Y for dark tones reduces ink consumption and drying time.

🔢 **RGB to CMYK Conversion (normalized RGB ∈ [0,1]):**

Let $R, G, B \in [0, 1]$. First, compute the black component:

$$
K = 1 - \max(R, G, B)
$$

Then compute the CMY components:

$$
C = \frac{1 - R - K}{1 - K} \quad
M = \frac{1 - G - K}{1 - K} \quad
Y = \frac{1 - B - K}{1 - K}
$$

Special case:

- If \( R = G = B = 0 \) (pure black), then \( C = M = Y = 0 \), \( K = 1 \)

✍️ **Notes:**

- CMYK is **device-dependent**, meaning values may vary depending on printer profiles and paper type.
- RGB → CMYK is **not reversible** in general, because CMYK has a smaller color gamut.
- Normalized values are often scaled to \([0, 100]\%\) or \([0, 255]\) for practical implementations.
- Pillow’s **default** conversion is effectively a **“Maximum GCR” (where GCR = 0)** approach. It treats *CMYK* as a simple 4-channel extension of *CMY*, where the $K$ channel is **redundant or ignored** unless specifically told to calculate it.

### <a id='toc3_3_4_'></a>[YUV (Luminance and Chrominance)](#toc0_)

A color space designed to **separate** luminance (brightness) information from chrominance (color) information.  

- **Y** represents the **luminance** component (**brightness**).  
- **U** and **V** represent the **chrominance** components (**color information**).
- It is generally from the **BT.601 standard**, which is commonly associated with standard-definition television.

💡 **Purpose of RGB to YUV Conversion:**

- **Bandwidth Optimization in Video:**  
  Since the human eye is more sensitive to brightness details than color details, YUV allows chrominance (U, V) to be subsampled, reducing the amount of color data transmitted or stored without a significant loss in perceived quality.

- **Compression:**  
  Many video and image compression standards (like MPEG, JPEG, and H.264) use YUV because it enables better compression efficiency by treating luminance and chrominance separately.

- **Broadcast and Transmission:**  
  YUV was originally developed for analog TV broadcasting to maintain backward compatibility with black-and-white televisions (which only needed the luminance signal).

- **Color Manipulation:**  
  Separating luminance and chrominance can simplify tasks like brightness adjustments and color filtering.

🔢 **RGB - YUV Conversion:**

$$
\begin{aligned}
\text{RGB → YUV:} \quad
& \begin{bmatrix} Y \\ U \\ V \end{bmatrix} =
\begin{bmatrix}
0.299 & 0.587 & 0.114 \\
-0.14713 & -0.28886 & 0.436 \\
0.615 & -0.51499 & -0.10001
\end{bmatrix}
\begin{bmatrix} R \\ G \\ B \end{bmatrix} \\
\text{YUV → RGB:} \quad
& \begin{bmatrix} R \\ G \\ B \end{bmatrix} =
\begin{bmatrix}
1 & 0 & 1.13983 \\
1 & -0.39465 & -0.58060 \\
1 & 2.03211 & 0
\end{bmatrix}
\begin{bmatrix} Y \\ U \\ V \end{bmatrix}
\end{aligned}
$$


#### <a id='toc3_3_4_1_'></a>[Manual](#toc0_)


```python
RGB_TO_YUV = np.array(
    [
        [0.29900, 0.58700, 0.11400],
        [-0.14713, -0.28886, 0.43600],
        [0.61500, -0.51499, -0.10001],
    ],
    dtype=np.float32,
)

YUV_TO_RGB = np.linalg.inv(RGB_TO_YUV)
YUV_OFFSET = np.array([0, 128, 128])
```

```python
def rgb2yuv(image: NDArray) -> NDArray[np.uint8]:
    img = (image.reshape(-1, 3) @ RGB_TO_YUV.T + YUV_OFFSET).reshape(image.shape)
    return np.clip(img, 0, 255).astype(np.uint8)


def yuv2rgb(image: NDArray) -> NDArray[np.uint8]:
    img = ((image.reshape(-1, 3) - YUV_OFFSET) @ YUV_TO_RGB.T).reshape(image.shape)
    return np.clip(img, 0, 255).astype(np.uint8)
```

### <a id='toc3_3_5_'></a>[YCbCr/YCC (Luminance and Chrominance)](#toc0_)

A color space used primarily in **digital video and image compression** (e.g., JPEG, MPEG, H.264), separating image luminance from chrominance components.

- **Y** represents the **luma** component (brightness approximation).
- **Cr** and **Cb** represent the **chrominance** components:
  - **Cb** is the blue-difference chroma component.
  - **Cr** is the red-difference chroma component.

💡 **Purpose of RGB to Y'CrCb Conversion:**

- **Digital Compression Efficiency:**  
  Y'CrCb allows for **chroma subsampling** (like 4:2:0), significantly reducing data with minimal perceptual loss due to human visual system sensitivity.

- **Standard in Digital Media:**  
  Widely used in standards like **JPEG**, **MPEG**, and **broadcast video**, where it allows efficient encoding and transmission.

- **Compatibility with Color TV Signals:**  
  While derived from analog YUV, YCrCb is tailored for **digital representation**, including defined value ranges and offsets.

🔢 **RGB - Y'CrCb Conversion:**

### <a id='toc3_3_6_'></a>[HSV/HSB (Hue, Saturation, Value/Brightness)](#toc0_)

- HSV stands for **Hue**, **Saturation**, and **Value** (or **Brightness** in HSB, which is functionally identical).
- It's a cylindrical color space designed to separate an image's intensity (Value/Brightness) from its color information (Hue and Saturation).
- This model is widely used in image processing, color pickers, and computer vision tasks due to its intuitive separation of color properties.

🪧 **Details:**

- **Hue (H):**
  - Represents the pure color type as an angle on the color wheel.
  - Typical range: [0°, 360°], though often scaled to [0, 1], [0, 255], or [0, 100] depending on the system.

- **Saturation (S):**
  - Measures the vibrancy or purity of the color.
  - 0% indicates a shade of gray (no color).
  - 100% indicates the full, pure color.
  - Typical ranges: [0, 1], [0, 255], or [0%, 100%].

- **Value (V) / Brightness (B):**
  - Indicates the overall brightness or lightness of the color.
  - 0% is black.
  - 100% is the brightest possible color.
  - Typical ranges: [0, 1], [0, 255], or [0%, 100%].

💡 **Purpose of RGB to HSV Conversion:**

- **Color Filtering and Detection:** HSV simplifies isolating specific colors because hue is largely independent of brightness and saturation variations.
- **Intuitive User Interfaces:** HSV is more aligned with human perception of color, making it ideal for color pickers and interactive design tools.
- **Image Editing and Enhancement:** Adjusting brightness or saturation independently of hue allows for more controlled and intuitive color manipulation.

🔢 **RGB to HSV Conversion (using normalized RGB where R, G, B ∈ [0,1]):**

Given $R, G, B \in [0, 1]$:

1. Find the maximum and minimum values:
    - $C_{\text{max}} = \max(R, G, B)$
    - $C_{\text{min}} = \min(R, G, B)$
    - $\Delta = C_{\text{max}} - C_{\text{min}}$

1. Calculate HSV components:

    $$
    V = C_{\text{max}}
    $$

    $$
    S = \begin{cases}
    0 & \text{if } C_{\text{max}} = 0 \\
    \frac{\Delta}{C_{\text{max}}} & \text{otherwise}
    \end{cases}
    $$

    $$
    H = \begin{cases}
    0 & \text{if } \Delta = 0 \\
    60^\circ \times \left( \left( \frac{G - B}{\Delta} \right) \pmod 6 \right) & \text{if } C_{\text{max}} = R \\
    60^\circ \times \left( \frac{B - R}{\Delta} + 2 \right) & \text{if } C_{\text{max}} = G \\
    60^\circ \times \left( \frac{R - G}{\Delta} + 4 \right) & \text{if } C_{\text{max}} = B
    \end{cases}
    $$

✍️ **Note on OpenCV Implementation:**

The OpenCV library scales these components differently when converting:
- **H:** Scaled to the range [0, 179] (to fit within a single 8-bit byte, as 180 × 2 = 360).
- **S and V:** Scaled to the range [0, 255].

### <a id='toc3_3_7_'></a>[HSL (Hue, Saturation, Lightness)](#toc0_)

- HSL stands for **Hue**, **Saturation**, and **Lightness**.
- Like HSV, it's a cylindrical color space designed to represent colors intuitively.
- Although it shares the Hue component with HSV, its Saturation and Lightness components differ, offering distinct advantages.

🪧 **Details:**

- **Hue (H):**
  - Identical to Hue in HSV, representing the pure color type as an angle on the color wheel.
  - Typical range: [0°, 360°], often scaled digitally to [0, 1], [0, 255], or [0, 100%].

- **Saturation (S):**
  - Represents the degree of colorfulness relative to its brightness.
  - Measures how much the color differs from a pure gray at the same lightness level.
  - 0% means the color is a shade of gray.
  - 100% means the color is fully saturated at its current lightness.
  - Typical ranges: [0, 1], [0, 255], or [0%, 100%].

- **Lightness (L):**
  - Represents the perceived brightness of the color, ranging from black to white.
  - 0% is black.
  - 100% is pure white.
  - 50% is a neutral, fully saturated color at its maximum colorfulness.
  - Typical ranges: [0, 1], [0, 255], or [0%, 100%].

💡 **Key Differences from HSV:**

The primary distinctions are in **Saturation** and **Lightness/Value**:

1. **Lightness (L) vs. Value (V):**
    - **HSV's Value (V):** Represents the *maximum* RGB component. Decreasing V makes the color darker by mixing it with black.
    - **HSL's Lightness (L):** Represents the *average* of the maximum and minimum RGB components. This results in more perceptually uniform brightness adjustments. L=0 is black, L=1 is white, and L=0.5 allows for the most "pure" colors.

2. **Saturation (S) Behavior:**
    - **HSV's Saturation:** Is dependent on Value. As Value decreases, colors can appear less saturated. A pure red (H=0, S=1, V=1) becomes a dark maroon as V lowers, even if S remains 1.
    - **HSL's Saturation:** Is independent of Lightness. Adjusting L changes the color's brightness without altering its fundamental "colorfulnes" relative to that lightness. A color at HSL(0, 100%, 50%) is pure red; changing L makes it darker or lighter but it remains "fully saturated" at that new lightness.

✍️ **When to Use Which?**

- **HSV:** Ideal for tasks requiring color detection, filtering based on intensity, or in computer vision where variations in brightness are key.
- **HSL:** Better suited for user interfaces, color pickers, and graphic design. It allows for more intuitive adjustments of lightness and darkness, creating consistent color palettes.

### <a id='toc3_3_8_'></a>[XYZ (CIE 1931)](#toc0_)

- **XYZ** is the **device-independent** color space defined by the **CIE** (*Commission Internationale de l’Éclairage*).
- Any visible color can be described using an XYZ triple `(X, Y, Z)` under a chosen reference illuminant/observer (commonly CIE 1931 2°).
- It’s not meant to be “intuitive” like **RGB**, but it’s the foundation for many other color spaces.

👁️ **How XYZ relates to human perception:**

- XYZ is built from *color matching functions* (human eye response curves).
- X, Y, Z are mathematical basis functions that make color representation convenient (not “red, green, blue” channels).
- Y is closely related to luminance (perceived brightness)

🔗 **Relation with Color Gamut Diagram:**

- To relate the XYZ coordinates to the color gamut diagram, we normalize the tristimulus values into chromaticity coordinates $(x, y, z)$.
- This process projects the 3D space into a 2D plane, separating chromaticity from luminance.

$$S = X + Y + Z$$

$$
x = \frac{X}{S}, \quad
y = \frac{Y}{S}, \quad
z = \frac{Z}{S}
$$

✍️ **Key properties of this relation:**

1. **Normalization:**
   - Because $x + y + z = 1$, any two coordinates define the third:
   $$z = 1 - x - y$$

1. **Luminance Independence:**
   - The $(x, y)$ coordinates describe the *chromaticity* (hue and saturation)
   - The $Y$ value from the original XYZ triple represents the *luminance* (brightness).

1. **The Gamut Diagram:** Plotting $y$ against $x$ on a Cartesian plane produces the CIE 1931 Chromaticity Diagram, where all visible colors are contained within the boundary of spectral colors.

### <a id='toc3_3_9_'></a>[LAB (Lightness, A, B)](#toc0_)

- Its primary goal is to be **perceptually uniform**, meaning that the distance between two colors in the L\*a\*b\* space corresponds to how humans perceive the difference between those colors.
- This makes it incredibly useful for color difference calculations, digital color management, and ensuring color consistency across different devices.
- Unlike RGB or CMYK, which are device-dependent, or HSV/HSL, which are related to RGB, CIELAB is designed to be device-independent.

🪧 **Details:**

- **Lightness (L):**
  - Represents the perceived lightness of a color.
  - It ranges from 0 (black) to 100 (white).
  - Mid-range values represent shades of gray.
  - Typical range: [0, 100].

- **a (Green-Red axis):**
  - This axis represents color information along a green-to-red scale.
  - Negative values indicate shades of green.
  - Positive values indicate shades of red.
  - A value of 0 means the color is on the neutral (gray) axis.
  - Typical range: Approximately [-128, +127] (though the exact bounds can vary slightly depending on the reference white point).

- **b (Blue-Yellow axis):**
  - This axis represents color information along a blue-to-yellow scale.
  - Negative values indicate shades of blue.
  - Positive values indicate shades of yellow.
  - A value of 0 means the color is on the neutral (gray) axis.
  - Typical range: Approximately [-128, +127] (similar to a\*).

## <a id='toc3_5_'></a>[Nonlinear Color and Photographic Effects](#toc0_)

Nonlinear color transformations modify pixel intensities using **nonlinear functions or mapping**, rather than simple linear matrices.  


### <a id='toc3_5_1_'></a>[Monochromatic Filters](#toc0_)

Monochromatic filters map an image to **shades of a single color** or grayscale while preserving relative intensity.  


#### <a id='toc3_5_1_1_'></a>[Duotone](#toc0_)


```python
def duotone(image: NDArray, color_dark: tuple[int, int, int], color_light: tuple[int, int, int]) -> NDArray:

    # convert to float
    img = image.astype(np.float32)

    # luminance (perceptual grayscale)
    gs = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gs = cv2.normalize(gs, None, 0, 1, cv2.NORM_MINMAX)  # normalize to [0, 1]

    # prepare colors
    c0 = np.array(color_dark, dtype=np.float32)
    c1 = np.array(color_light, dtype=np.float32)

    # nonlinear interpolation
    duotone_img = (1 - gs[..., None]) * c0 + gs[..., None] * c1

    return np.clip(duotone_img, 0, 255).astype(np.uint8)
```

```python
duotone_1 = duotone(img_3_np, color_dark=(20, 30, 100), color_light=(150, 30, 20))
duotone_2 = duotone(img_3_np, color_dark=(40, 30, 100), color_light=(0, 200, 0))
duotone_3 = duotone(img_3_np, color_dark=(50, 200, 30), color_light=(150, 0, 150))

# plot
fig, axes = plt.subplots(1, 4, figsize=(16, 4), layout="compressed")
titles = ["Original", "Duotone 1", "Duotone 2", "Duotone 3"]
images = [img_3_np, duotone_1, duotone_2, duotone_3]

for ax, img, title in zip(axes, images, titles):
    ax.imshow(img, vmin=0, vmax=255)
    ax.set_title(title)
    ax.axis("off")
plt.show()
```

#### <a id='toc3_5_1_2_'></a>[Tritone](#toc0_)


```python
def tritone(
    image: NDArray,
    color_shadow: tuple[int, int, int],
    color_mid: tuple[int, int, int],
    color_high: tuple[int, int, int],
) -> NDArray:

    # convert to float
    img = image.astype(np.float32)

    # luminance (perceptual grayscale)
    gs = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gs = cv2.normalize(gs, None, 0, 1, cv2.NORM_MINMAX)  # normalize to [0, 1]

    # prepare colors
    c0 = np.array(color_shadow, dtype=np.float32)
    c1 = np.array(color_mid, dtype=np.float32)
    c2 = np.array(color_high, dtype=np.float32)

    out = np.zeros_like(img)

    # shadow to midtones
    mask_low = gs <= 0.5
    t = gs[mask_low] * 2.0
    out[mask_low] = (1 - t[:, None]) * c0 + t[:, None] * c1

    # midtones to highlight
    mask_high = gs > 0.5
    t = (gs[mask_high] - 0.5) * 2.0
    out[mask_high] = (1 - t[:, None]) * c1 + t[:, None] * c2

    return np.clip(out, 0, 255).astype(np.uint8)
```

```python
tritone_1 = tritone(
    img_3_np,
    color_shadow=(20, 30, 120),  # deep blue shadows
    color_mid=(180, 160, 80),  # warm midtones
    color_high=(250, 240, 220),  # soft highlights
)
tritone_2 = tritone(
    img_3_np,
    color_shadow=(60, 40, 20),  # dark brown
    color_mid=(160, 120, 80),  # sepia midtones
    color_high=(240, 220, 180),  # parchment highlights
)
tritone_3 = tritone(
    img_3_np,
    color_shadow=(0, 0, 0),  # black
    color_mid=(120, 120, 120),  # neutral gray
    color_high=(255, 255, 255),  # white
)

# plot
fig, axes = plt.subplots(1, 4, figsize=(16, 4), layout="compressed")
titles = ["Original", "Tritone 1", "Tritone 2", "Tritone 3"]
images = [img_3_np, tritone_1, tritone_2, tritone_3]

for ax, img, title in zip(axes, images, titles):
    ax.imshow(img, vmin=0, vmax=255)
    ax.set_title(title)
    ax.axis("off")
plt.show()
```
